```python
path = "/Users/dominick/Microscopy-AI-Agent-Demo/artifacts/2026-03-23_21-13-47/scan_211355_512x512.npy"
```

```python
from pathlib import Path

import matplotlib.pyplot as plt

import numpy as np

data = np.load(path, allow_pickle=True)

name = Path(path).name



print(f"Loaded: {name}")

print(f"Shape: {data.shape}")

print(f"Dtype: {data.dtype}")



if data.ndim == 0:

    print("Scalar value:", data.item())

elif data.ndim == 1:

    plt.figure(figsize=(10, 4))

    plt.plot(data)

    plt.title(Path(path).stem)

    plt.xlabel("Index")

    plt.ylabel("Value")

    plt.tight_layout()

    plt.show()

elif data.ndim == 2:

    plt.figure(figsize=(8, 8))

    plt.imshow(data, cmap="gray")

    plt.title(Path(path).stem)

    plt.colorbar()

    plt.axis("off")

    plt.tight_layout()

    plt.show()

elif data.ndim == 3 and data.shape[-1] in (3, 4):

    plt.figure(figsize=(8, 8))

    plt.imshow(data)

    plt.title(Path(path).stem)

    plt.axis("off")

    plt.tight_layout()

    plt.show()

else:

    raise ValueError(f"Unsupported array shape for display: {data.shape}")
```
